## Exercise 1

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='purple', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='X vs Y')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(["Y values with errors"], loc='lower right')
plt.show()
```

## Exercise 2

```python
x = np.linspace(0,50,500)
np_x = np.array(x)
sin_x = np.sin(np_x)
e_index = - (np_x / 10)
np_y = np.e ** e_index * sin_x

plt.plot(x, np_y, 
         color='steelblue')
plt.xlabel('x', fontsize = 15, color = "darkorange")
plt.ylabel('y', fontsize = 15, color = "darkorange")
plt.ylim([-1.00, 1.00])
plt.title('$y = e^{-x/10}sin(x)$', fontsize = 20)
plt.grid(alpha=.25)
plt.show()
```
